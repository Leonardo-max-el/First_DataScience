Gestion de la informacion duplicado o faltante, si el dato no nos interesa lo habitual es eliminarlos

```python
#Modificaci√≥n de la estrucutura de nuestra base de datos
import pandas as pd
df = pd.read_csv('C:/Users/Jhoana FE/Desktop/datasets/base_datos_2008.csv', nrows=1e6)
print(df.columns)
```

```python
df.columns = df.columns.str.strip()  # Elimina espacios en blanco
print(df.columns.tolist())  # Verifica nuevamente
```

```python
#Duplicamos los datos con el mismo dataframe
dfduplicate = pd.concat([df, df])
```

```python
#Reordenamos el dataframe
dfduplicate = dfduplicate.sample(frac=1)
```

```python
#Eliminamos los casos duplicados
dfclean = dfduplicate.drop_duplicates()
```

```python
#Comprobamos si nuestro nuevo df es igual al df original 
len(dfclean) == len(df)
```
