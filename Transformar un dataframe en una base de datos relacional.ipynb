Vamos a ver un ejemplo práctico de cómo transformar una base de datos clásica, de las que podemos encontrar en cualquier hoja de cálculo de Excel, por ejemplo, en una estructura de datos relacional. El método que voy a mostrar aquí no es único en absoluto, pero te va a permitir entender cuál es la idea detrás de las transformaciones que realizamos. 

```python
import pandas as pd

data = [
    [1, "Carlos", "Ramírez", 28, 101, "Laptop Dell XPS 13", 4200, 3500, 700, 1, "2025-04-20"],
    [1, "Carlos", "Ramírez", 28, 101, "Laptop Dell XPS 13", 4200, 3500, 700, 1, "2025-04-20"],
    [1, "Carlos", "Ramírez", 28, 101, "Laptop Dell XPS 13", 4200, 3500, 700, 1, "2025-04-20"],
    [2, "Lucía", "Fernández", 32, 102, "Teclado Mecánico Redragon", 250, 180, 70, 2, "2025-04-18"],
    [2, "Lucía", "Fernández", 32, 102, "Teclado Mecánico Redragon", 250, 180, 70, 2, "2025-04-18"],
    [3, "Andrés", "Gutiérrez", 30, 103, "Mouse Logitech MX Master 3", 320, 250, 70, 1, "2025-04-17"],
    [4, "María", "Pérez", 26, 104, "Monitor Samsung 27'' 4K", 1350, 1100, 250, 1, "2025-04-15"],
    [5, "Javier", "Morales", 35, 105, "Audífonos Sony WH-1000XM5", 950, 800, 150, 3, "2025-04-10"]
]
```

```python
labels = [
    "Comprador_id", "Nombre", "Apellido", "Edad", "Producto_id", "Producto",
    "Precio", "Coste", "Margen", "Cantidad", "Fecha"
]
```

```python
df = pd.DataFrame.from_records(data, columns=labels)
```

```python
df
```

```python
#Dividiremos la tablas en otras para tener informacion requerida
compradores = df.drop_duplicates(subset = "Comprador_id", keep = "first")
compradores
```

```python
#Obtenemos las tablas que queremos
compradores = compradores[["Comprador_id", "Nombre", "Apellido","Edad"]]
compradores
```

```python
productos = df.drop_duplicates(subset = "Producto_id", keep="first" )
```

```python
productos = productos [["Producto_id","Producto","Precio","Coste","Margen"]]
productos
```
